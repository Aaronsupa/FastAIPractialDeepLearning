```python
from fastai.vision.all import *
from fastai.text.all import *
from fastai.collab import *
from fastai.tabular.all import *
```

```python
path = untar_data(URLs.ADULT_SAMPLE)
dbls = TabularDataLoaders.from_csv(path/'adult.csv', path = path, y_names = "salary", 
        cat_names = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "education-num_na"],
        cont_names = ["age", "fnlwgt", "education-num"],
        procs = [Categorify, FillMissing, Normalize],)
dbls.show_batch()
```

```python
learn = tabular_learner(dbls, metrics=accuracy)
learn.fit_one_cycle(2)
```

```python
learn.show_results()
```

```python
df = pd.read_csv(path/'adult.csv')
df.head()
```

```python
splits = RandomSplitter(valid_pct=0.2)(range_of(df))

to = TabularPandas(df, procs=[Categorify, FillMissing,Normalize],
                   cat_names = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race'],
                   cont_names = ['age', 'fnlwgt', 'education-num'],
                   y_names='salary',
                   splits=splits)

dls = to.dataloaders(bs=64)

dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(1)
learn.show_results()
```

```python
row, clas, probs = learn.predict(df.iloc[0])
row.show()
```
